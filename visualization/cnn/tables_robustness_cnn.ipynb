# Functions for creating tables

## Untargeted Attacks

```python
def create_table_untargeted_mean(decimals: int = 2):
    table = pd.DataFrame(columns = [['normal', 'normal', 'adversarial', 'adversarial'], 
                                    ['PRS', 'RMSE [%]', 'PRS', 'RMSE [%]']],
                         index=['No attack', 'PGD', 'Noise'])
    
    table.loc['No attack'] = [None,
                              results_pgd_untargeted_normal_training['RMSE_Unmodified_Target_Original'].mean() * 100,
                              None,
                              results_pgd_untargeted_adversarial_training['RMSE_Unmodified_Target_Original'].mean() * 100,]
    
    table.loc['PGD'] = [results_pgd_untargeted_normal_training['PRS'].mean(),
                        results_pgd_untargeted_normal_training['RMSE_Attacked_Target_Original'].mean() * 100,
                        results_pgd_untargeted_adversarial_training['PRS'].mean(),
                        results_pgd_untargeted_adversarial_training['RMSE_Attacked_Target_Original'].mean() * 100,]
    
    table.loc['Noise'] = [results_noise_normal_training['PRS'].mean(),
                          results_noise_normal_training['RMSE_Attacked_Target_Original'].mean() * 100,
                          results_noise_adversarial_training['PRS'].mean(),
                          results_noise_adversarial_training['RMSE_Attacked_Target_Original'].mean() * 100,]
    
    return table.astype('float').round(decimals)
```

```python
def create_table_untargeted_std(decimals: int = 2):
    table = pd.DataFrame(columns = [['normal', 'normal', 'adversarial', 'adversarial'], 
                                    ['PRS', 'RMSE [%]', 'PRS', 'RMSE [%]']],
                         index=['No attack', 'PGD', 'Noise'])
    
    table.loc['No attack'] = [None,
                              results_pgd_untargeted_normal_training['RMSE_Unmodified_Target_Original'].std() * 100,
                              None,
                              results_pgd_untargeted_adversarial_training['RMSE_Unmodified_Target_Original'].std() * 100,]
    
    table.loc['PGD'] = [results_pgd_untargeted_normal_training['PRS'].std(),
                        results_pgd_untargeted_normal_training['RMSE_Attacked_Target_Original'].std() * 100,
                        results_pgd_untargeted_adversarial_training['PRS'].std(),
                        results_pgd_untargeted_adversarial_training['RMSE_Attacked_Target_Original'].std() * 100,]
    
    table.loc['Noise'] = [results_noise_normal_training['PRS'].std(),
                          results_noise_normal_training['RMSE_Attacked_Target_Original'].std() * 100,
                          results_noise_adversarial_training['PRS'].std(),
                          results_noise_adversarial_training['RMSE_Attacked_Target_Original'].std() * 100,]
    
    return table.astype('float').round(decimals)
```
